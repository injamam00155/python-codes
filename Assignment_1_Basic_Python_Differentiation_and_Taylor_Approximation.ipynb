a) Generate values for a list **x** from 0 to 2 with iterval of 0.1.

b) Find the **2nd order Taylor approximation** for the function :

> $$y = e^{sin(x)}$$

around, 

> $$x_0 = 0$$

c) Use the Taylor approximation to define a variable as: 
>$$ y\_taylor = taylor\ approximaiton\ eq^n\ of\ [y = e^{sin(x)}]\ around\ x_0 = 0$$
*Note: the taylor approximation should be defined in terms of the x variable defined before*

d) Plot a graph with **x** on the x-axis and the corresponding **y_taylor** values on the y-axis.

*The plot should look similar to the figure below*



image.png

````python
#code here
x=np.arange(0,2,0.1)
def y_t(x):
  return np.exp(np.sin(x))
y=y_t(x)
print(x)
print(y)
def y_d1(x):
  return np.exp(np.sin(x))*np.cos(x)
def y_d2(x):
  return np.exp(np.sin(x))*np.cos(x)**2-np.exp(np.sin(x))*np.sin(x)
def y_taylor(x,a):
  return y_t(a)+y_d1(a)*(x-a)+0.5*y_d2(a)*(x-a)**2
plt.plot(x,y_taylor(x,0))
````
